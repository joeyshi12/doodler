```python
img = Image.open("images/goku.png")
draw = ImageDraw.Draw(img)
font = ImageFont.truetype("static/fonts/sauce_code_pro.ttf", 50)
rows, cols = img.size

draw.text((5, 5), "Peter Sux", font=font, align="left", fill="black")
img.resize((rows // 4, cols // 4))
```
